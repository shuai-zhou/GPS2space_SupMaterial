# Load libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from matplotlib.patches import Patch

import datetime
import numpy as np
import pandas as pd
import geopandas as gpd
from geopandas import GeoSeries
from shapely.geometry import Point

from gps2space import geodf, space, dist

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

# Make data

## Generate random data

```python
xmin, xmax, ymin, ymax = -80.52, -80.50, 39.50, 39.52
```

```python
np.random.seed(123)
x_a = (xmax - xmin) * np.random.random(20) + xmin
y_a = (ymax - ymin) * np.random.random(20) + ymin
person_a = gpd.GeoDataFrame(GeoSeries([Point(x, y) for x, y in zip(x_a, y_a)]))
person_a.rename(columns={0:'geometry'}, inplace=True)
person_a['id'] = 'person_a'
person_a['timestamp'] = pd.to_datetime('2021-05-12')
person_a['day'] = person_a['timestamp'].dt.day
```

```python
np.random.seed(1234)
x_b = (xmax - xmin) * np.random.random(30) + xmin
y_b = (ymax - ymin) * np.random.random(30) + ymin
person_b = gpd.GeoDataFrame(GeoSeries([Point(x, y) for x, y in zip(x_b, y_b)]))
person_b.rename(columns={0:'geometry'}, inplace=True)
person_b['id'] = 'person_b'
person_b['timestamp'] = pd.to_datetime('2021-05-12')
person_b['day'] = person_b['timestamp'].dt.day
```

## Separate lat/long

```python
person_a['longitude'] = person_a['geometry'].x
person_a['latitude'] = person_a['geometry'].y
person_a[['id','timestamp','longitude','latitude','day']].to_csv('./data/person_a.csv', index=False)
```

```python
person_b['longitude'] = person_b['geometry'].x
person_b['latitude'] = person_b['geometry'].y
person_b[['id','timestamp','longitude','latitude','day']].to_csv('./data/person_b.csv', index=False)
```

# Example I
**Buffer- and Convex hull-based activity space and shared space**

## Convert raw GPS data into spatial data

```python
df_a = pd.read_csv('./data/person_a.csv')
df_b = pd.read_csv('./data/person_b.csv')
```

```python
df_a.head()
df_b.head()
```

```python
gdf_a = geodf.df_to_gdf(df_a, x='longitude', y='latitude')
gdf_b = geodf.df_to_gdf(df_b, x='longitude', y='latitude')
```

```python
gdf_a.head()
gdf_b.head()
```

## Spatial projection and spatial measure extraction

```python
gdf_a = gdf_a.to_crs('epsg:2163')
gdf_b = gdf_b.to_crs('epsg:2163')
```

```python
buff_a = space.buffer_space(gdf_a, dist=500, dissolve='day', proj=2163)
buff_b = space.buffer_space(gdf_b, dist=500, dissolve='day', proj=2163)
convex_a = space.convex_space(gdf_a, group='day', proj=2163)
convex_b = space.convex_space(gdf_b, group='day', proj=2163)
```

```python
buff_a
buff_b
convex_a
convex_b
```

## Extraction of shared space by overlapping activity space

```python
buff_share = gpd.overlay(buff_a, buff_b, how='intersection')
buff_share['share_space'] = buff_share['geometry'].area

convex_share = gpd.overlay(convex_a, convex_b, how='intersection')
convex_share['share_space'] = convex_share['geometry'].area
```

```python
buff_share
convex_share
```

## Plot activity space and shared space

```python
buffa_mi = (buff_a['buff_area'] * 0.0000003861).values[0]
buffb_mi = (buff_b['buff_area'] * 0.0000003861).values[0]
convexa_mi = (convex_a['convex_area'] * 0.0000003861).values[0]
convexb_mi = (convex_b['convex_area'] * 0.0000003861).values[0]
buffshare = (buff_share['share_space'] * 0.0000003861).values[0]
convexshare = (convex_share['share_space'] * 0.0000003861).values[0]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(25, 15))

gdf_a.plot(ax=ax1, marker='o', markersize=8, c='g')
gdf_b.plot(ax=ax1, marker='^', markersize=8, c='r')
buff_a.boundary.plot(ax=ax1, edgecolor='g', linewidth=2)
buff_b.boundary.plot(ax=ax1, edgecolor='r', linewidth=2)
buff_share.boundary.plot(ax=ax1, facecolor='grey', alpha=0.4)

ax1.axes.xaxis.set_visible(False)
ax1.axes.yaxis.set_visible(False)
ax1.set_title('(a) Buffer-based activity space\n and shared space', fontsize=20)

gdf_a.plot(ax=ax2, marker='o', markersize=8, c='g')
gdf_b.plot(ax=ax2, marker='^', markersize=8, c='r')
convex_a.boundary.plot(ax=ax2, edgecolor='g', linewidth=2)
convex_b.boundary.plot(ax=ax2, edgecolor='r', linewidth=2)
convex_share.boundary.plot(ax=ax2, facecolor='grey', alpha=0.6)

ax2.axes.xaxis.set_visible(False)
ax2.axes.yaxis.set_visible(False)
ax2.set_title('(b) Convex hull-based activity space\n and shared space', fontsize=20)

leg_ax1 = [Line2D([0],[0], marker='o', c='w',
                  label='Geolocation_Person A', mfc='g', markersize=8),
           Line2D([0],[0], marker='^', c='w',
                  label='Geolocation_Person B', mfc='r', markersize=8),
           Line2D([0],[0], c='green', lw=2,
                  label='Activity space_Person A ({0:.2f} $mi^{{2}}$)'.format(buffa_mi)),
           Line2D([0],[0], c='red', lw=2,
                  label='Activity space_Person B ({0:.2f} $mi^{{2}}$)'.format(buffb_mi)),
           Patch(facecolor='grey', alpha=0.6,
                 label='Shared space ({0:.2f} $mi^{{2}}$)'.format(buffshare))]
leg_ax2 = [Line2D([0],[0], marker='o', c='w',
                  label='Geolocation_Person A', mfc='g', markersize=8),
           Line2D([0],[0], marker='^', c='w',
                  label='Geolocation_Person B', mfc='r', markersize=8),
           Line2D([0],[0], c='green', lw=2,
                  label='Activity space_Person A ({0:.2f} $mi^{{2}}$)'.format(convexa_mi)),
           Line2D([0],[0], c='red', lw=2,
                  label='Activity space_Person B ({0:.2f} $mi^{{2}}$)'.format(convexb_mi)),
           Patch(facecolor='grey', alpha=0.6,
                 label='Shared space ({0:.2f} $mi^{{2}}$)'.format(convexshare))]

ax1.legend(handles=leg_ax1, loc='lower left', fontsize=11, frameon=False)
ax2.legend(handles=leg_ax2, loc='lower left', fontsize=11, frameon=False)

fig.subplots_adjust(wspace=0.0, hspace=0.0)
plt.savefig('./figure/Figure_2.png', bbox_inches='tight', dpi=600)
```
